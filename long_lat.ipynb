## Pulling LAT/LONG from google API

```python
import pandas as pd
from geopy.geocoders import GoogleV3
```

```python
# Create a geocoder instance
geolocator = GoogleV3(api_key='AIzaSyDSOG9YGnJC8qLtvb-rDnDNgn2zRoJ93so')
```

```python
data = pd.read_excel('/Volumes/T7/Career/School/MSADSB/BTA 520/housing_normal.xlsx')
```

```python
# Define functions to get latitude and longitude
def get_latitude(zipcode):
    location = geolocator.geocode(zipcode)
    return location.latitude if location else None

def get_longitude(zipcode):
    location = geolocator.geocode(zipcode)
    return location.longitude if location else None
```

```python
# Add latitude and longitude columns to the DataFrame
data['latitude'] = data['zipcode'].apply(get_latitude)
data['longitude'] = data['zipcode'].apply(get_longitude)
```
