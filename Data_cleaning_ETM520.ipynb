```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None
```

```python
#read in data
housing_df = pd.read_csv("/Users/jwitch/Desktop/Dumps/portland_housing.csv")
```

```python
housing_df
```

```python
#shape and dtypes of data
print(housing_df.shape)
print(housing_df.dtypes)
```

```python
cols = housing_df.columns[:20] # first 30 columns
colours = ['#000099', '#ffff00'] # specify the colours - yellow is missing. blue is not missing.
sns.heatmap(housing_df[cols].isnull(), cmap=sns.color_palette(colours))
```

```python
for col in housing_df.columns:
    pct_missing = np.mean(housing_df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
housing_df_updated = housing_df[['address/zipcode', 'price', 'lastSoldPrice', 'homeStatus', 'yearBuilt', 'bathrooms',
                                  'bedrooms', 'daysOnZillow', 'zestimate', 'homeType',
                                  'resoFacts/roofType', 'lastSoldPrice', 'livingArea', 'priceHistory/1/price', 'propertyTaxRate', 'latitude', 'longitude']]
```

```python
cols = housing_df_updated.columns[:20] # first 30 columns
colours = ['#000099', '#ffff00'] # specify the colours - yellow is missing. blue is not missing.
sns.heatmap(housing_df[cols].isnull(), cmap=sns.color_palette(colours))
```

```python
clean_df = housing_df_updated.dropna(how= 'any', axis=0)
clean_df
```

```python
cols = clean_df.columns[:20] # first 30 columns
colours = ['#000099', '#ffff00'] # specify the colours - yellow is missing. blue is not missing.
sns.heatmap(clean_df[cols].isnull(), cmap=sns.color_palette(colours))
```
